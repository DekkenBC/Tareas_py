```python
from asyncio.windows_events import NULL #Da error en Colab pero funciona en Visual Studio
import pprint
import random
from operator import itemgetter

def main():
    Trabajadores = []
    IDs = []
    while True:
        cont = len(Trabajadores)
        print ("Base de datos de trabajadores de la empresa")
        print("1.- Agregar automatico")
        print("2.- Agregar manual")
        print("3.- Imprimir")
        print("4.- Buscar por ID")
        print("5.- Ordenar")
        print("6.- Eliminar por ID")
        print("7.- Borrar toda la lista")
        print("0.- Salir")
        try:
            op = int(input("Su seleccion: "))
        except ValueError:
            print("Seleccione un numero valido")
        else:
            match op:
                case 1:
                    print("")
                    ND,NI = act1(IDs)
                    Trabajadores.append(ND)
                    IDs.append(NI)
                    print("Usuario creado con exito\n")

                case 2:
                    print("")
                    ND,NI = act2(IDs)
                    Trabajadores.append(ND)
                    IDs.append(NI)
                    print("Usuario creado con exito\n")

                case 3:
                    print("")
                    if len(Trabajadores) == 0:
                        print("No hay datos que mostrar")
                    for n in Trabajadores:
                        pprint.pprint(n,sort_dicts=False)
                        print("")
                    print("")

                case 4:
                    print("")
                    indice = buscar(Trabajadores, cont)
                    if indice == NULL:
                        print("No se ha encontrado a este trabajador")
                    elif indice == "NoHay":
                        print("No hay trabajadores actualmente")
                    elif indice == None:
                        print("Se ha producido un error inesperado")
                    else:
                        indice = indice - 1
                        pprint.pprint(Trabajadores[indice],sort_dicts=False)
                    print("")

                case 5:
                    print("")
                    Trabajadores.sort(key = itemgetter("Clave"))
                    #Orden = sorted(Trabajadores, key = itemgetter("Clave"))
                    #Trabajadores = Orden
                    print("Los trabajadores se han ordenado con exito")
                    print("")

                case 6:
                    print("")
                    indice = buscar(Trabajadores, cont)
                    if indice == NULL:
                        print("No se ha encontrado a este trabajador")
                    elif indice == "NoHay":
                        print("No hay trabajadores actualmente")
                    elif indice == None:
                        print("Se ha producido un error inesperado")
                    else:
                        indice = indice - 1
                        Trabajadores.remove(Trabajadores[indice])
                        print("Trabajador eliminado correctamente")
                    print("")

                case 7:
                    print("")
                    seguro = input("Seguro que desea borrar la base de datos? (Si/No)")
                    if seguro == "Si":
                        Trabajadores.clear()
                    else:
                        print("Uff, que bien, me diste un gran susto")
                    print("")
                case 0:
                    break

                case _:
                    print("Seleccione un numero valido")

def act1(Claves):
    nombresM = ["Carlos","Juan","Pedro","Alejandro","Alexis","Paul","Rodrigo","Rogelio","Daniel","Angel"]
    nombresF = ["Alexis","Damaris","Alejandra","Karla","Dayana","Paula","Daniela","Marisa","Sara","Dora"]
    nombresO = nombresM + nombresF
    Apellidos = ["Hernandez","Perez","Huerta","Bermudez","Lopez","Gonzales","Rodriguez","Sanchez","Ramirez"]

    while True:
        Clave = random.randint(1,500)
        if Clave in Claves:
            pass
        else:
            break

    select = random.randint(1,3)
    if select == 1:
        Genero = "Masculino"
    elif select == 2:
        Genero = "Femenino"
    else:
        Genero = "Otro"

    if Genero == 1:
        Nombre = random.choice(nombresM)
    if Genero == 2:
        Nombre = random.choice(nombresF)
    else:
        Nombre = random.choice(nombresO)

    ApellidoPat = random.choice(Apellidos)
    ApellidoMat = random.choice(Apellidos)

    Crear_Auto = {'Clave':Clave,'Nombre':Nombre,'Apellido Paterno':ApellidoPat,'Apellido Materno':ApellidoMat,'Genero':Genero}

    return Crear_Auto, Clave

def act2(Claves):
    while True:
        try:
            Clave = int(input("Dame la clave de trabajador: "))
        except ValueError:
            print ("Seleccione un numero valido")
        else:
            if Clave < 1:
                print("Seleccione un numero positivo mayor a 0")
            elif Clave in Claves:
                print("Esta ID ya esta en uso")
            else:
                break

    Nombre = input("Dame el nombre del trabajador: ")
    Apellido_Pat = input("Dame el apellido materno del trabajador: ")
    Apellido_Mat = input("Dame el apellido paterno del trabajador: ")

    while True:
        Genero = input("Dame el genero del trabajador (Masculino, Femenino, Otro): ")
        if Genero == "Masculino" or Genero == "Femenino" or Genero == "Otro":
            break
        else:
            print("Seleccione uno de los generos especificados\n")

    Crear = {'Clave':Clave,'Nombre':Nombre,'Apellido Paterno':Apellido_Pat,'Apellido Materno':Apellido_Mat,'Genero':Genero}

    return Crear, Clave

def buscar(Lista, maxi):
    stop = 0
    cont = 1
    if maxi == 0:
        return "NoHay"
    while True:
        try:
            buscar = int(input("Cual es el ID?: "))
        except ValueError:
            print("Error en la busqueda, seleccione un numero entero mayor a 0")
        else:
            if buscar < 1:
                return NULL
            while True:
                for x in Lista:
                    if cont == maxi + 1:
                        return NULL
                    for key,value in x.items():
                        if buscar == value:
                            return(cont)
                        else:
                            cont = cont + 1
                            break


main()
```
