```python
import random

def main():
    while True:
        print("Actividades")
        print("1.- Listas de mascotas o artistas, mostrarlos y el numero de letras en cada palabra")
        print("2.- Generar numeros y mostrar lista")
        print("3.- Sumar listas")
        print("4.- Eliminar elementos duplicados de una lista")
        print("5.- Calcular media y mediana de una lista")
        print("6.- Salir")

        try:
            op = int(input("Su seleccion: "))
        except ValueError:
            print("Seleccione una opcion valida\n")
        else:
            match op:
                case 1:
                    print("")
                    act1()
                    print("")
                case 2:
                    print("")
                    act2()
                    print("")
                case 3:
                    print("")
                    act3()
                    print("")
                case 4:
                    print("")
                    act4()
                    print("")
                case 5:
                    print("")
                    act5()
                    print("")
                case 6:
                    break
                case _:
                    print("Seleccione una opcion valida\n")


def act1():
   lista = []
   for nom in range(10):
       nombre = input("Dame el nombre de un artista o mascota (Minimo 5 y precionar 0 para salir): ")
       if nom >= 5 and nombre == "0":
           break
       else:
           lista.append(nombre)

   for x in range(len(lista)):
       print(f"[{x+1}] {lista[x]} - {len(lista[x])} caracteres")

def act2():
    numeros = []

    while len(numeros) < 10:
        for x in range(10):
            num = random.randint(30,50)
            if num in numeros:
                pass
            elif len(numeros) == 10:
                break
            else:
                numeros.append(num)

    print(list(enumerate(numeros,1)))

def act3():
    lista1 = []
    lista2 = []
    lista3 = []

    for i in range(2):
        for x in range(10):
            try:
                num = int(input(f"Dame un numero para la lista {i+1} (0 para terminar): "))
            except ValueError:
                print("Use solo numeros por favor")
            else:
                if num == 0:
                    break
                if i == 0:
                    lista1.append(num)
                else:
                    lista2.append(num)

    if len(lista1) < len(lista2):

        print("La primera lista es mas corta que la segunda")
        for a in range(len(lista1)):
            lista3.append(lista1[a] + lista2[a])
    else:
        if len(lista1) > len(lista2):
            print("La segunda lista es mas corta que la primera")
        for a in range(len(lista2)):
            lista3.append(lista1[a] + lista2[a])

    print(f"primera lista = {lista1}")
    print(f"segunda lista = {lista2}")
    print(f"listas sumadas = {lista3}")

def act4():
    numeros = []
    numeros2 = []

    while True:
        try:
            num = int(input("Dame un numero para la lista (0 para salir): "))
        except ValueError:
            print("Solo se aceptan numeros\n")
        else:
            if num == 0:
                break
            else:
                numeros.append(num)

    for x in numeros:
        if x not in numeros2:
            numeros2.append(x)

    print(numeros2)

def act5():
    lista1 = []
    ind = 0

    while True:
        try:
            num = int(input("Dame un numero para la lista (0 para terminar): "))
        except ValueError:
            print("Use solo numeros por favor")
        else:
            if num == 0:
                break
            else:
                lista1.append(num)

    media = sum(lista1) / len(lista1)
    print(f"La media de la lista es: {media:.2f}")
    lista1.sort()

    if len(lista1)%2 == 0:
        ind = int(len(lista1)/2)
        num1 = lista1[ind-1]
        num2 = lista1[(ind)]
        print(lista1)
        print(f"Las medianas de la lista fueron {num1} y  {num2}")

    else:
        lista1.append(0)
        ind = int(len(lista1)/2)
        mediana = lista1[ind-1]
        lista1.remove(0)
        print(lista1)
        print(f"La mediana de la lista es {mediana}")

if __name__ == "__main__":
    main()
```
